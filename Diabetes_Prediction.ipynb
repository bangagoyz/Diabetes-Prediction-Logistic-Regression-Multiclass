#Diabetes Prediction Using Logistic Regression Multiple Class

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

```python
diabetes_data = pd.read_csv('Dataset of Diabetes.csv')
```

```python
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
diabetes_data.isna().sum()
```

```python
diabetes_data['CLASS'].value_counts()
```

```python
diabetes_data['Gender'] = diabetes_data['Gender'].str.upper() #Uppercase all Gender datas
```

Y = Diabetic

N = Non Diabetic

P = Predict Diabetic

```python
diabetes_data['CLASS'] = diabetes_data['CLASS'].str.strip() #REMOVE WHITE SPACE
diabetes_data.drop(columns=['ID', 'No_Pation'], inplace=True) #Remove Unused Feature
```

```python
diabetes_data['Gender'] = diabetes_data['Gender'].map({'F': 0, 'M':1}) #Change string into number
```

```python
diabetes_data.groupby('CLASS').mean()
```

```python
X = diabetes_data.drop(columns='CLASS', axis=0)
Y = diabetes_data['CLASS']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=42, stratify=Y)
```

```python
sm = SMOTE()
X_res, Y_res = sm.fit_resample(X_train, Y_train)
```

```python
X_res.shape
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=5000, class_weight='balanced')
```

```python
print(X_train)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy training data:', training_data_accuracy)
print('Accuracy test data:', test_data_accuracy)
```

```python
print(" Predicted Labels:")
print(X_test_prediction)
```

```python
print("Accuracy:", accuracy_score(Y_test, X_test_prediction))
print("\nClassification Report:\n", classification_report(Y_test, X_test_prediction))
```
